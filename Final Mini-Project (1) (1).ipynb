```python
from tkinter import ttk
import csv
from tkinter import *
from tkinter import messagebox
users=[]
info=[]
gk=[] 
#--------------------------------------------------------------------------------------------------------------------------
#--------------------------------------------------------------------------------------------------------------------------
def validatepass(l5,l6,ques,ans,b3):
    p=pw.get()
    c=cpw.get()
    if p!=c:
        messagebox.showwarning(title='Warning',message='Recheck password')
        return False
    #print("ee")
    l5.pack()
    ques.pack()
    l6.pack()
    ans.pack()
    b3.pack()
def createuser():
        t=[]
        t.append(uid.get())
        t.append(pw.get())
        t.append(ques.get())
        t.append(ans.get())
        users.append(t)
        with open('infotest.csv','a', newline='') as file:
            writer=csv.writer(file)
            for i in users:
                writer.writerow(i)
        messagebox.showinfo(title='Welcome',message='User created successfully')
    
def dosignin():
    global L
    flag=False
    uid=luid.get()
    pwd=lpwd.get()
    #print("Entered")
    with open('infotest.csv','r', newline='') as file:
        csvreader = csv.reader(file,delimiter=',')
        for row in csvreader:
            if uid==row[0]:
                #print("Enter1")
                if pwd==row[1]:
                    #print("Enter2")
                    messagebox.showinfo(title='Welcome',message='Welcome '+row[0]) 
                    flag=True
                    gkgame()
                else:
                    messagebox.showwarning(title='Error',message='Invalid  password')
                    yn=messagebox.askyesno(title='Error',message='Do you want recover?')
                    if yn==1:
                        l9.pack()
                        lseq.pack()
                        lseq.insert(0,str(row[2]))
                        l10.pack()
                        lans.pack()
                        b5.pack()
                        L=row
                        print(row)
                        flag=True
            if flag==False:
                 messagebox.showwarning(title='Error',message='Record not Found')
def showpass():
    global L
    if lans.get()==L[3]:
        print('Enter 1')
        passen.pack()
        passen.insert(0,L[1])
    else:
        messagebox.showinfo(title='Error',message='Wrong answer')
    
#-------------------------------------------------------------------------------------------------------------------------                
#--------------------------------------------------------------------------------------------------------------------------
                    
def gkgame():
    gk=[]
    hst=[]
    def start():
        value=val.get()
        global r
        global score
        r,score=0,0
        if value==1:
            r2.pack_forget()
            with open("gktest.csv", 'r') as file: #comma separator version-csv
                csvreader = csv.reader(file,delimiter=',')
                for row in csvreader:
                     gk.append(row)
                q.insert(0,gk[r][0])
                #print(gk)
        else:
            r1.pack_forget()
            with open("histtest.csv", 'r') as file: #comma separator version-csv
                csvreader = csv.reader(file,delimiter=',')
                for row in csvreader:
                     hst.append(row)
                q.insert(0,hst[r][0])
                #print(hst)
    def Next(x):
        global r
        global score
        value=val.get()
        q.delete(0,'end')
        ans.delete(0,'end')
        if value==1:
            if(x!=1):
                r+=1
            if r>=len(gk):
                messagebox.showinfo('',message='Your score is '+ str(score))
                messagebox.showinfo("","End of Exam")
                page1.destroy()
                page.destroy()
            else:
                q.insert(0,gk[r][0])
                    
        else:
            if(x!=1):
                r+=1
            if r>=len(hst):
                messagebox.showinfo('',message='Your score is '+ str(score))
                messagebox.showinfo("","End of Exam")
                page1.destroy()
                page.destroy()
            else:
                q.insert(0,hst[r][0])
                
    def save():
        global r
        global score
        value=val.get()
        if value==1:
            if ans.get()==gk[r][1]:
                messagebox.showinfo("","Correct")
                score+=1
            else:
                messagebox.showinfo("","Incorrect")
            r+=1
            if r>=len(gk):
                messagebox.showinfo('',message='Your score is '+ str(score))
                messagebox.showinfo("","End of Exam")
                page1.destroy()
                page.destroy()
            else:
                #print("ee")
                Next(1)
        else:
            if ans.get()==hst[r][1]:
                messagebox.showinfo("","Correct")
                score+=1
            else:
                messagebox.showinfo("","Incorrect")
            r+=1
            if r>=len(hst):
                messagebox.showinfo('',message='Your score is '+ str(score))
                messagebox.showinfo("","End of Exam")
                page1.destroy()
                page.destroy()
            else:
                Next(1)
    def exit():
        global score
        messagebox.showinfo('',message='Your score is '+ str(score))
        page1.destroy()
        page.destroy()
    page1=Toplevel()
    page1.title('New File')
    page1.geometry('500x500')
    l=Label(page1,text="GK Quiz",font=('Times New Roman',30)).pack()
    l0=Label(page1,text='Select the Type of Questions: ').pack()
    val=IntVar()
    r1=Radiobutton(page1,text="Science Related ",variable=val,value=1)
    r1.pack()
    r2=Radiobutton(page1,text="History Related",variable=val,value=2)
    r2.pack()
    b4=Button(page1,text='Start',command=start)
    b4.pack()
    l1=Label(page1,text='Question').pack()
    q=Entry(page1)
    q.pack()
    l2=Label(page1,text=' Enter your Answer').pack()
    ans=Entry(page1)
    ans.pack()
    b1=Button(page1,text='Submit',command=save)
    b1.pack()
    b2=Button(page1,text='Next',command=lambda:Next(0))
    b2.pack()
    b3=Button(page1,text='End Test',command=exit)
    b3.pack()
    page1.mainloop()
#--------------------------------------------------------------------------------------------------------------------------
#--------------------------------------------------------------------------------------------------------------------------                    
page=Tk()
page.geometry('500x500')
LOGIN = ttk.Notebook(page)
LOGIN.pack(pady=10, expand=True)
signup = ttk.Frame(LOGIN, width=400, height=280)
signin = ttk.Frame(LOGIN, width=400, height=280)

signup.pack(fill='both', expand=True)
signin.pack(fill='both', expand=True)


LOGIN.add(signup, text='Register/signup as new user')
LOGIN.add(signin, text='Login/signin')



l1=Label(signup,text='Enter Full Name: ').pack()
uid=Entry(signup)
uid.pack()
l3=Label(signup,text="Password").pack()
pw=Entry(signup,show='*')
pw.pack()
l4=Label(signup,text="Confirm password:").pack()
cpw=Entry(signup,show='*')
cpw.pack()
b2=Button(signup,text="Validate password",command=lambda: validatepass(l5,l6,ques,ans,b3)).pack()
b3=Button(signup,text="Register User",command=createuser)


l5=Label(signup,text="Enter security ques")
ques=Entry(signup)
l6=Label(signup,text="Answer")
ans=Entry(signup)
b3=Button(signup,text="Register User",command=createuser)


l7=Label(signin,text='User ID').pack()
luid=Entry(signin)
luid.pack()
l8=Label(signin,text="Password").pack()
lpwd=Entry(signin,show='*')
lpwd.pack()

b4=Button(signin,text="Signin",command=dosignin)
b4.pack()
l9=Label(signin,text="security question is: ")
lseq=Entry(signin)
l10=Label(signin,text="please enter your answer")
lans=Entry(signin)
b5=Button(signin,text="Show password",command=showpass)
passen=Entry(signin)

page.mainloop()

```
